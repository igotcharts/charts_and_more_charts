```python
os.chdir(outputs)
line = charts_function_list.chart_maker()

fig = line.initial_fig_axis(figsize=(8,5))
mpl.rcParams['font.family'] = 'Futura Bk BT'
fig.suptitle('Charity Events Planned at Mar-a-Lago',size=22)
ax = line.axes_set_up(fig=fig)
ax.plot(range(len(count)),count, color='#f03b20',linewidth=3,zorder=2,alpha=.8,marker='o',markersize=)

ax.set_xticks(range(5))
ax2 = ax.twinx()
ax2.plot(range(len(count)),[1 for x in range(5)], color='#2c7fb8',linewidth=3,zorder=2,alpha=.8)
ax2.set_ylim(0,1.1)
ax2.set_yticks([0,1])
ax2.set_yticklabels(['Not Racist','Racist'],color='#2c7fb8')


ax.set_xticklabels(year)
ax.set_xlabel('Years',fontsize=20,alpha=.8)


for item,color,label,label_pad in ((ax,'#f03b20','Events Planned',10),(ax2,'#2c7fb8','Trump\'s Racism Level',-30)):
    line.tick_params_(item,pad=15,fontsize=16,labelcolor=color)
    item.set_ylabel(label,size=20,color=color,labelpad=label_pad,alpha=.8)
ax.grid(color='grey',alpha=.1,linewidth=2)
line.patch_adder(ax,facecolor='#f0f0f0')
line.citations(ax,x=-.1,source='Source: David Fahrenthold\'s twitter - @Fahrenthold',source_y=-.38,chart_tag_y=-.44,fontsize=10)
#charts_function_list.chart_save('mar_a_lago_drop')
plt.show()
```
