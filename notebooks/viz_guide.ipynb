```python
def qualities(i,title):
    ax = fig.add_subplot(2,4,i,title=title)
    ax.title.set_size(14)
    ax.tick_params(labelsize=12)
    return ax
```

```python
fig = plt.figure(figsize=(11,6))
fig.suptitle(t='Basic Chart Types',size=20)
fig.subplots_adjust(top=.85)
ax1 = qualities(1,'Line Chart')
ax1.plot([1,2,3],[6,8,12])

ax2 = qualities(2,'Bar Chart')
ax2.bar([1,2,3],[6,8,12])

ax3 = qualities(3,'Horizontal Bar Chart')
ax3.barh([1,2,3],[6,8,12])

ax4 = qualities(4,'Pie Chart')
ax4.pie([2,7,1],colors=['Blue','Green','White'])
ax4.scatter(0,0,marker='x',s=10000,color='red',zorder=2)


ax5 = qualities(5,'Stacked Bar')
ax5.bar([1,2,3],[2,6,8])
ax5.bar([1,2,3],[8,4,2],bottom=[2,6,8],color='orange')
ax5.set_yticklabels([str(x)+'%' for x in list(range(0,101,25))])

ax6 = qualities(6,'Scatter Plot')
ax6.scatter(np.random.randint(0,100,size=40),np.random.randint(0,100,size=40),s=4,color='red',zorder=2)
ax6.set_xticks([0,25,50,75,100])
ax6.set_yticks([0,25,50,75,100])

ax7 = qualities(7,'Map')
ax7.grid(False)
ax7.set_xticks([])
ax7.set_yticks([])

ax8 = qualities(8,'Box Plot')
ax8.boxplot(x=[np.random.normal(size=100),np.random.exponential(size=50),np.random.rand(100)])

plt.show()
```
