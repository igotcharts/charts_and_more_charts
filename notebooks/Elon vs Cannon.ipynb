```python
cannon_elon = pd.read_csv('cannon vs elon.csv')
```

```python
plt.rcParams['font.family'] = 'Trebuchet MS' 
fig, ax = plt.subplots(figsize=(11,8))
fig.suptitle('Pronatalist Battle Royale: Elon vs Nick',fontsize=20)
plt.plot(cannon_elon['Year'],cannon_elon['Cannon'],label='Nick Cannon',color='#1b9e77')
plt.scatter(x=cannon_elon['Year'],y=cannon_elon['Cannon Year'],facecolors='none',edgecolor='#1b9e77',s=25)
plt.plot(cannon_elon['Year'],cannon_elon['Elon'],label='Elon Musk',color='#7570b3')
plt.scatter(x=cannon_elon['Year'],y=cannon_elon['Elon Year'],facecolors='none',edgecolor='#7570b3',s=25)
ax.set_xlabel('Year',fontsize=14)
ax.set_ylabel('Kids',fontsize=14)
ax.text(0,-.1,'@igotcharts|2025',transform=ax.transAxes,color='#d95f02')
ax.set_xlim(1999.8,2025.15)
ax.legend(bbox_to_anchor=(.19,.95),loc=9)
plt.savefig('elon_nick.png',bbox_inches = 'tight', dpi = 300, pad_inches = .5,edgecolor = '#ffffff',facecolor = '#ffffff')
```
